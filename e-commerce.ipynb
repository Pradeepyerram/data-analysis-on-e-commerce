```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing data
tx=pd.read_csv('ecommerce_transactions.csv')
cust=pd.read_csv('customers_profile.csv')
prod=pd.read_csv('products_catalog.csv')
```

```python
tx.head()
```

```python
tx.groupby('product_id')['revenue'].sum().sort_values(ascending=False).head(10)
```

```python
# Merge customers with transactions on 'customer_id'
customer_transactions = pd.merge(tx, cust, on='customer_id', how='left')
```

```python
# Merge the above result with products on 'product_id'
full_data = pd.merge(customer_transactions, prod, on='product_id', how='left')
```
